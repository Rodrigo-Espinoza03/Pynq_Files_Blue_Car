```python
from pynq.overlays.base import BaseOverlay
from pynq.lib import Pmod_PWM
import time

# Load the base overlay
base = BaseOverlay("base.bit")

# Set up the PWM on PMODA pin 0
pwm = Pmod_PWM(base.PMODA, 0)

# Set the PWM period to 20 ms (20,000 us for 50 Hz)
period = 20000  # 20 ms period in microseconds

# Function to set servo position with pulse width in microseconds
def set_servo_pulse(pulse_width_us):
    # Convert pulse width to percentage for the duty cycle (out of 20 ms)
    duty_cycle = (pulse_width_us / period) * 100
    pwm.generate(period, int(duty_cycle))

try:
    while True:
        if base.buttons[0].read():
            # Move servo to hard right position (2400 us pulse width)
            set_servo_pulse(2400)
        elif base.buttons[1].read():
            # Move servo to neutral position (1500 us pulse width)
            set_servo_pulse(1500)
        elif base.buttons[2].read():
            # Move servo to minimum position (600 us pulse width)
            set_servo_pulse(600)
        elif base.buttons[3].read():
            # Move servo to hard left position (600 us pulse width)
            set_servo_pulse(600)
        else:
            # Return servo to neutral position if no button is pressed
            set_servo_pulse(1500)

        time.sleep(0.1)  # Small delay to avoid overwhelming the CPU

except KeyboardInterrupt:
    pwm.stop()
    print("PWM stopped")
```
